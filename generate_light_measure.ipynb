```python
import os
import random
import time
from datetime import datetime
import pandas as pd

# Zielverzeichnis
output_dir = "sensor_data/csv"
os.makedirs(output_dir, exist_ok=True)

# Funktion zur Generierung zuflliger Lichtmesswerte
def generate_light_data(sensor_id=2):
    dt_now = datetime.now().isoformat()
    return {
        "sensor_id": sensor_id,
        "timestamp": dt_now,
        "light_level_lux": round(random.uniform(0, 1000), 2),  # 01000 Lux realistisch fr Innenrume
        "day_of_week": random.choice([
            "monday", "tuesday", "wednesday",
            "thursday", "friday", "saturday", "sunday"
        ])
    }

# Benutzerdefinierte Anzahl
num_iterations = int(input("Enter number of light sensor readings to generate: "))

# Daten generieren
light_readings = []
for i in range(num_iterations):
    light_data = generate_light_data()
    print(light_data)
    light_readings.append(light_data)
    time.sleep(0.01)

    # Optional: Duplikat bei jedem 3. Eintrag
    if i % 3 == 0 and i != 0:
        light_readings.append(light_data)
        print(light_data)

# DataFrame erstellen
df = pd.DataFrame(light_readings)

# Duplikate entfernen
df = df.drop_duplicates()

# Fehlende Werte entfernen (falls ntig)
df = df.dropna()

# CSV speichern
csv_path = os.path.join(output_dir, "light_sensor.csv")
df.to_csv(csv_path, index=False)
print(f"Saved light sensor data to CSV: {csv_path}")
```
